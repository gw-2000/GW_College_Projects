### Question 1: Load Datasets

Details of the performances by the New York Philharmonic orchestra are contained in the files **concerts.csv**, **works.csv**, and **soloists.csv**. Each row within **concerts.csv** describes a performance, using the followiung fields:

* **ID**: The ID of the performance
* **Event_Type**: The type of the performance
* **Location**: The location of the performance
* **Country**: The country where the performance was held
* **Venue**: The venue of the performance
* **Date**: The date of the performance, given as dd/mm/yyyy (e.g. 24/07/2014, 02/03/2012)
* **Start_Time**: The time that the performance started given as hh:mm (e.g. 17:30, 11:00)
* **End_Time**: The time that the performance ended given as hh:mm (e.g. 17:30, 11:00)
* **Duration**: The duration of the concert

Each row within **works.csv** contains details of the works performed in a particular performance, using the following fields:

* **ID**: The ID of the performance (links to the data in the concerts.csv file)
* **Orchestra**: The name of the orchestra that performed
* **ComposerName**: The name of the composer(s) whose work was performed
* **ConductorName**: The name of the conductor

Each row within **soloists.csv** contains details of the soloist(s) that performed in a particular performance, using the following fields:

* **ID**: The ID of the performance (links to the data in the concerts.csv file)
* **Soloist_Rank**: The rank of the solist within in the perfomrance (e.g. 1st, 2nd, 3rd, ...)
* **Soloist_Name**: The name of the soloist
* **Soloist_Instrument**: The instrument the soloist played

Load the three datasets (**concerts.csv**, **works.csv**, and **soloists.csv**) into pandas data frames: **concerts**, **works**, and **soloists**. Display the first five rows from each.

### Question 3: Deeper Analysis

Use slightly more advanced data analysis to answer the following questions.

**a)** Draw an appropriate data visualsiation that shows the number of concerts played in each location outside of the United Statess of America?

```python
non_us = concerts.loc[concerts['Country'] != 'USA']
non_us_loc = non_us.Location
loc_freq = non_us.Location.value_counts()

loc_visual = pd.DataFrame({'Frequency': loc_freq})
ax = loc_visual.plot.bar(y='Frequency', align='center',rot=45, figsize=(20,10), 
                         title = "Locations Performed Outside the US:")
ax
```

**b)** How many times has there been a concert anywhere in Germany?

**Hint:** Experiment with the *str.contains* method from *pandas.Series*.

```python
ger_sr = pd.Series(concerts.Country)
german_shows = ger_sr[ger_sr.str.contains('Germany', case=False, regex=True)]

print("The orchestra has performed in Germany on", len(german_shows), "occasions.")
```

**c)** What is the average duration of a concert (in minutes)?

```python
avg_dur = concerts.Duration.mean()
avg_dur_rounded = round(avg_dur)

print("The average duration of a concert is", avg_dur_rounded, "minutes, or", avg_dur, "to be exact!")
```

-----------

**d)** Are performances of works by Beethoven of longer duration, on average, than performances of works by Mozart? 

**Bonus:** Can you plot a data visualisation to support this conclusion?

```python
comp_dur= pd.concat([works.ComposerName, concerts.Duration], axis=1)
comp_dur.dropna(inplace=True)

beet_works_avg = comp_dur[comp_dur['ComposerName'].str.contains("Beethoven", case=False, regex=True)]
moz_works_avg = comp_dur[comp_dur['ComposerName'].str.contains("Mozart", case=False, regex=True)]

beet_avg = round(beet_works_avg.Duration.mean())
moz_avg = round(moz_works_avg.Duration.mean())

print(" On average, performances of works by Mozart are longer in duration at", moz_avg, "minutes", 
      "compared with Beethoven's\n", beet_avg, "minutes. See the bar chart below:")

avg_visual = pd.DataFrame({'Composer':['Beethoven', 'Mozart'], 'Average Duration': [beet_avg, moz_avg]})
my_colors = list(['xkcd:faded red', 'xkcd:light navy'])
ax_avg = avg_visual.plot.bar(x='Composer', y='Average Duration', align='center',rot=0, figsize=(10,5), 
                         title = "Average Duration of each Composer's Performance:", 
                             ylim=(beet_avg-3,moz_avg+3), color=my_colors, legend=False)

```

**e)** Determine the most popular composer for each year in the dataset? 

```python
ID_join = works["ID"] + concerts["ID"]
ID_join.drop_duplicates()

pop_comp_yr = pd.concat([ID_join, works["ComposerName"], concerts["Date"]], axis=1)
pop_comp_yr.dropna(inplace=True)

pop_comp_yr["Date"] = pop_comp_yr.Date.str.slice(-4)
year_data = pop_comp_yr["Date"].drop_duplicates()


year_16 = pop_comp_yr[(pop_comp_yr.Date == '2016')]
year_16 = year_16.ComposerName.value_counts()[:1]
year_15 = pop_comp_yr[(pop_comp_yr.Date == '2015')]
year_15 = year_15.ComposerName.value_counts()[:1]
year_14 = pop_comp_yr[(pop_comp_yr.Date == '2014')]
year_14 = year_14.ComposerName.value_counts()[:1]
year_13 = pop_comp_yr[(pop_comp_yr.Date == '2013')]
year_13 = year_13.ComposerName.value_counts()[:1]
year_12 = pop_comp_yr[(pop_comp_yr.Date == '2012')]
year_12 = year_12.ComposerName.value_counts()[:1]
year_11 = pop_comp_yr[(pop_comp_yr.Date == '2011')]
year_11 = year_11.ComposerName.value_counts()[:1]

d = {'Year': year_data, 'Composer Name': [year_11.to_string(), year_12.to_string(), year_13.to_string(), year_14.to_string(),
                                          year_15.to_string(), year_16.to_string()]}
df = pd.DataFrame(data=d)
df

```
